# 4. What method takes a list argument and writes it to a 
# CSV file?

```python
# ANS: 4 
# The csv. writer.writerow(list) method takes a list or any (iterable) as an 
# argument. and writes it to a .csv file
import csv
# The list to be written to .csv  file
Fruits = ['apple','cherry','orange','pineapple','strawberry']
# Open the output file in 'w' mode
with open('fruits.csv','w') as f:
    # Create a csv.writer object with excel dialect
    wr = csv.writer(f, dialect='excel')
    # use writerow(List) of csv writer object created above to write the list to file
    wr.writerow(Fruits)
# The image below shows .CSV file created and list is writtwn as row
```

# 5. What do the keyword arguments delimiter and line 
# terminator do?

```python
# ANS 5
# The delimiter and line terminator arguments are second and third 
# arguments of csv.writer() function. delimiter is the character that appears
# between cells on a row. 
# By default, the delimiter for a CSV file is a comma.
# The line terminator is the character that comes at the end of a row. 
# By default, the line terminator is a newline. 
# We can change characters to different values by using the delimiter and 
# lineterminator keyword arguments with csv.writer()
```

```python
import csv
csvFile = open('output.tsv', 'w', newline='') # .tsv means tab separated value
csvWriter = csv.writer(csvFile, delimiter='\t', lineterminator='\n\n')
# Note that delimiter is changed from default "," to a TAB character
csvWriter.writerow(['word1', 'word2', 'word3'])
csvWriter.writerow(['word4', 'word5', 'word6'])
csvWriter.writerow(['word7', 'word8', 'word9', 'word10', 'word11', 'word12'])
csvFile.close()
# The image below shows output filewhere words are separated by TAB
```

# 6. What function takes a string of JSON data and returns
# a Python data structure?

```python
# ANS: 6
# conversion of JSON objects into their respective Python objects is called 
# deserialization and json.load() function is used for it
```

```python
import json
json_var ="""
{
    "Country": {
        "name": "INDIA",
        "Languages_spoken": [
            {
                "names": ["Hindi", "English", "Bengali", "Telugu"]
            }
        ]
    }
}
"""
print(type(json_var))
output = json.loads(json_var)
print(output)
print(type(output))
# Thus json.loads() takes a json string as input and gives a dictionary 
# as output.
```

# 7. What function takes a Python data structure and returns
# a string of JSON data?

```python
# ANS:7 json.dump() function takes a python data structure as argument and 
# returns a json data. This process is called serialization
```

```python
# E.g. using above code in reverse direction
import json
python_dict = {'Country': {'name': 'INDIA', 'Languages_spoken': [{'names': ['Hindi', 'English', 'Bengali', 'Telugu']}]}}
# Create an output file to store json data
with open("output.json", "w") as f:
    json.dump(python_dict, f)
# The image below shows that python dictioanry is successfuly written in file
# output.json
```
