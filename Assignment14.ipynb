# 1. What does RGBA stand for?

```python
# ANS: 1
# Computer programs often represent a color in an image as an RGBA value. An
# RGBA value is a group of numbers that specify the amount of red, green, blue,
# and alpha (or transparency) in a color. Each of these component values is an 
# integer from 0 (none at all) to 255 (the maximum). These RGBA values are 
# assigned to individual pixels. RGBA values are shown by four-tuples. E.g.
# Red: (255, 0, 0, 255)
# Green:(0,255,0,255)
# Blue:(0,0,255,255)
# White:(255,255,255,255)
# Black:(0,0,0,255)
```

# 3. What is a box tuple, and how does it work?

```python
# ANS: 3
# Some Pillowâ€™s functions take a box tuple argument. which is a tuple of four
# integer coordinates that represent a rectangular region in an image. The four
# integers are, in order, as follows:
# Left: The x-coordinate of the leftmost edge of the box.
# Top: The y-coordinate of the top edge of the box.
# Right: The x-coordinate of one pixel to the right of the rightmost edge of 
# the box. This integer must be greater than the left integer.
# Bottom: The y-coordinate of one pixel lower than the bottom edge of the box.
# This integer must be greater than the top integer.
# For example, the image below is shown by box tuple (3,1,9,6)=(L,T,R,B)
# Left = 3, Top = 1, Right = 9, Bottom = 5
```

# 5. What method would you call to get Image object for a 100Ã—100 image, excluding the lower-left quarter of it?

```python
# We can call self defined function get_image(): as belwo
# The image is shown above
from PIL import Image
import numpy as np    
def get_image(path):
    # Opening the image and converting 
    # it to RGB color mode
    img = Image.open(path).convert('RGB')
    # Extracting the image data &
    # creating an numpy array out of it
    img_arr = np.array(img)
    # Turning the pixel values of the 400x400 pixels to black 
    img_arr[75:100,0:25] = (0, 0, 0) # last 0 stands for black

    # # Creating an image out of the previously modified array
    img = Image.fromarray(img_arr)
    img.save("Cropped_Image.png")
get_image(r"hundred_hundred.png")
# Result image is shown below
```

# 7. What module contains Pillowâ€™s shape-drawing code?

```python
# ANS7:
# ImageDraw module of the Python image processing library Pillow (PIL) provides
# methods for drawing shapes, such as circles, squares, and straight lines.
```

```python
from PIL import Image, ImageDraw
im = Image.new('RGB', (500, 300), (128, 128, 128)) # frame to store images
draw = ImageDraw.Draw(im)
draw.ellipse((0, 0, 200, 200), fill=(0, 0, 255), outline=(0, 0, 0))
draw.ellipse((100, 100, 150, 200), fill=(255, 0, 0), outline=(0, 0, 0))
draw.rectangle((200, 100, 300, 200), fill=(0, 192, 192), outline=(255, 255, 255))
draw.line((350, 200, 450, 100), fill=(255, 255, 0), width=10)
im.save('pillow_imagedraw.jpg', quality=90)
```
